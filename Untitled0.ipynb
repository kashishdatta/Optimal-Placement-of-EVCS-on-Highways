```python
import matplotlib.pyplot as plt

# Data
years = [2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030]
predicted_stations = [3, 10, 33, 120, 428, 1419, 3763, 6770, 8621, 9305, 9508]

# Create the graph
fig, ax = plt.subplots(figsize=(12, 8))

# Create the bars for predicted charging stations
ax.bar(years, predicted_stations, color='#428085', width=0.6)

# Add labels and title
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Number of Charging Stations', fontsize=14)
ax.set_title('Predicted Number of Charging Stations (2020-2030)', fontsize=16)

# Set the x-axis tick labels and positions
plt.xticks(years, fontsize=12, rotation=45)

# Remove the grid
ax.grid(False)

# Format the y-axis tick labels
ax.tick_params(axis='y', labelsize=12)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))

# Add borders to the graph
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

# Adjust the layout and display the graph
fig.tight_layout()
plt.show()
```

```python
import pandas as pd

data1 = pd.read_csv('/content/simulation_results2.csv')
```

```python
# @title Total Time (minutes) vs Number of Stops

from matplotlib import pyplot as plt
data1.plot(kind='scatter', x='Total Time (minutes)', y='Number of Stops', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Car ID vs Total Time (minutes)

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Car ID']
  ys = series['Total Time (minutes)']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = data1.sort_values('Car ID', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Car ID')
_ = plt.ylabel('Total Time (minutes)')
```
